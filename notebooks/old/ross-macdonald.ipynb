```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import integrate
```

```python
alpha = 0.1
beta = 0.15
gamma = 0.05
N = 200
I0 = alpha * N
S0 = N - I0

tmax = 60
t = np.arange(tmax)
sns.set_theme()
```

```python
def derivative(X, t):
    S, I = X
    dotS = -beta * S * I / N - gamma * alpha * S + gamma * (1 - alpha) * I
    dotI = - dotS
    return np.array([dotS, dotI])
```

```python
X0 = S0, I0 #Initial conditions vector
res = integrate.odeint(derivative, X0, t)
S, I = res.T
```

```python
plt.figure()
# plt.title("odeint method")
plt.plot(t, I, label='Infected')
plt.plot(t, S, label='Susceptible')
plt.xlabel('Timesteps')
plt.ylabel('Count')
plt.ylim([0,N])
plt.legend()
plt.title("Deterministic Susceptible-Infected Model with Migration")
plt.show();
```
